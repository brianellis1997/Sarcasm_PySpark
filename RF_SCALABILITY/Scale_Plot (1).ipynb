```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('execution_times_rf.csv')

# Plot Execution Time vs. Number of Nodes for each unique number of tasks per node
plt.figure(figsize=(10, 6))
for tasks_per_node in df[' tasks_per_node'].unique():
    subset = df[df[' tasks_per_node'] == tasks_per_node]
    plt.plot(subset['nodes'], subset[' execution_time'], marker='o', label=f'Tasks per Node: {tasks_per_node}')

plt.title('Execution Time vs. Number of Nodes')
plt.xlabel('Number of Nodes')
plt.ylabel('Execution Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()

# Plot Execution Time vs. Tasks per Node for each unique number of nodes
plt.figure(figsize=(10, 6))
for nodes in df['nodes'].unique():
    subset = df[df['nodes'] == nodes]
    plt.plot(subset[' tasks_per_node'], subset[' execution_time'], marker='o', label=f'Nodes: {nodes}')

plt.title('Execution Time vs. Tasks per Node')
plt.xlabel('Tasks per Node')
plt.ylabel('Execution Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()

"""
First Plot (Execution Time vs. Number of Nodes): This plot visualizes how the execution time changes with the number of nodes for each specified number of tasks per node. It helps you understand the scalability of your application as you increase the computing resources.

Second Plot (Execution Time vs. Tasks per Node): This plot shows how the execution time varies with the number of tasks assigned to each node, for different numbers of nodes. This can provide insights into how efficiently your application parallelizes tasks across cores within a node.

These plots will allow you to analyze the performance implications of distributing your PySpark tasks across different numbers of nodes and different task distributions within those nodes, helping you to fine-tune your application's scalability and efficiency.
"""
```
